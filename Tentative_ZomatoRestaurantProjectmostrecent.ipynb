```python
#Applying zscore
from scipy.stats import zscore
out_features=data1[['Average Cost for two','Votes']] 
z=np.abs(zscore(out_features))
np.where(z>2.9)
```

```python
z.iloc[20,1]      # To confirm that value at row 20, and column 1 is greater than 3
```

```python
# Now removing the data above zscore and creating new DataFrame so that original df may preserve and could be used for IQR method
df3=data1[(z<2.9).all(axis=1)]
```

```python
# calculate dataloss percentage between data1 and df1
print('data loss percentage:', ((data1.shape[0] - df3.shape[0]) / data1.shape[0]) * 100)
```

```python
df3.skew()
```

```python
# Now we can apply IQR method on original df

Q1 = out_features.quantile(0.25)
Q3 = out_features.quantile(0.75)
IQR = Q3 - Q1

# Correcting the filtering condition
df4 = data1[~((data1 < (Q1 - 1.5 * IQR)) | (data1 > (Q3 + 1.5 * IQR))).any(axis=1)]
df4.shape
```

```python
df4.skew()
```

```python
# calculate dataloss percentage between data1 and df4
print('data loss percentage:', ((data1.shape[0] - df4.shape[0]) / data1.shape[0]) * 100)
```

```python
# Though IQR handles skewness better than zscore but it results in significant data loss
# so combining zscore with cbrt can be a reasonable solution for this kind of data considering Aggregate raitng also
```

```python
df3[['Average Cost for two','Votes','Aggregate rating']]=np.cbrt(df3[['Average Cost for two','Votes','Aggregate rating']])
```

```python
# Function to repeatedly check accuracy of different models

def metric_score(pred, X_train, X_test, Y_train, Y_test, train=True):
    if train:
        Y_pred_train = pred.predict(X_train)
        print('\n============== Train Results ==============')
        print(f'r2_score: {r2_score(Y_train, Y_pred_train) * 100:.2f}%')
       
    else:
        Y_pred_test = pred.predict(X_test)
        print("\n============== Test Results ==============\n")
        print(f'r2_score: {r2_score(Y_test, Y_pred_test) * 100:.2f}%')
        print(f'Mean_Absolute_Error: {mean_absolute_error(Y_test, Y_pred_test):.2f}%')
        print(f'Mean_Squared_Error: {mean_squared_error(Y_test, Y_pred_test):.2f}%')
        print(f'Root_Mean_Squared_Error: {np.sqrt(mean_squared_error(Y_test, Y_pred_test)):.2f}%')
```
